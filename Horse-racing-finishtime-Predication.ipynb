# Detecting and removing outliers by "racekey" feature 

```python
## Make racekey feature using DATE , pzonecode and RNO(race number)

df_2['racekey'] = df_2['DATE'].astype(str)+"~"+df_2['pzonecode_x'].astype(str)+"~"+df_2['RNO'].astype(str)
filter_data = df_2[df_2.racekey=='20230215~US-NORTHFIELD_PARK-HS~1']
```

```python
## The process of finding racewise outliers 

racekey_col = "racekey"
features = df_2.select_dtypes(include=['number']).columns.tolist()
```

```python
Q1 = df_2[features].quantile(0.25)
Q3 = df_2[features].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = (df_2[features] < lower) | (df_2[features] > upper)
```

```python
df_2["Total_Outliers"] = outliers.sum(axis=1)
```

```python
df_2["RNO_Outlier_Percentage"]= (df_2["Total_Outliers"] / len(features)) * 100
h1= df_2[[racekey_col, "Total_Outliers", "RNO_Outlier_Percentage"]]
h1
```

```python
racekey_outliers = df_2.groupby("racekey")["Total_Outliers"].sum()
total_data_points = df_2.groupby("racekey").size() * len(features)  

racekey_outlier_percentage = (racekey_outliers / total_data_points) * 100
racekey_outlier_percentage
```

```python
p1 = racekey_outlier_percentage.reset_index()
p1.columns = ["racekey", "Racekey_Outlier_Percentage"]

p1
```

```python
## Visulation of racekey wise outliers  

plt.figure(figsize=(11, 6))
sns.histplot(p1['Racekey_Outlier_Percentage'], bins=120, color="purple" , kde=True , edgecolor="black" , alpha=0.7)
plt.xlabel("Outliers percentage per racekey", fontsize=14)
plt.ylabel("Frequency" , fontsize=14)
plt.title("Positive Distribution of racekey_wise_outliers" , fontsize=18)
plt.grid(axis='y' , linestyle="--" , alpha=0.6 , color="black")
plt.show()
```

```python
p2 = p1[p1["Racekey_Outlier_Percentage"] <= 9]
p2
```

```python
## Understanding data after filtering outliers above 10%

plt.figure(figsize=(11, 6))
sns.histplot(p2['Racekey_Outlier_Percentage'], bins=60, color="black" , kde=True , edgecolor="black" , alpha=0.7)
plt.xlabel("Outliers percentage per racekey", fontsize=14)
plt.ylabel("Frequency" , fontsize=14)
plt.title("Positive Distribution of racekey_wise_outliers" , fontsize=18)
plt.grid(axis='y' , linestyle="--" , alpha=0.6 , color="black")
plt.show()
```

```python
## finally removing outliers above 9%

df_2_cleaned = df_2[~df_2["racekey"].isin(p1[p1["Racekey_Outlier_Percentage"] >= 9]["racekey"])]
df_2_cleaned
```

```python
drop_data = ['racekey' , 'RNO' , 'DATE' , 'pzonecode_x' , 'Total_Outliers' , 'RNO_Outlier_Percentage']
```

```python
## it is final reuslt 

final_df = df_2_cleaned.drop(columns = drop_data)
final_df
```
